## 1.2 Basic preprocessing

**Stop word removal and lowercasing**

```python
stop_words = set(stopwords.words("english"))
no_stop_words = [token.lower() for token in moby_dick if not token.lower() in stop_words]
print(no_stop_words[0:10])
```

**Punctuation removal**

```python
filtered_moby = [token for token in no_stop_words if token.isalpha()]
print(filtered_moby[0:6])
```

**Both operations can be combined**

```python
filtered_moby = [token.lower() for token in moby_dick if not token.lower() in stop_words and not token in string.punctuation]
```

```python
print(len(set(filtered_moby))/len(filtered_moby))
```

## 1.4 Basic context

**Find the ten most/least common words**

## 1.6 Wordcloud

**Initialize word cloud**

```python
moby_dick_str = " ".join(filtered_moby)
wordcloud = WordCloud(width=1600, height=800, background_color="white", min_font_size=10).generate(moby_dick_str)
```

**Create the Word cloud**

```python
plt.figure(figsize=(16, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
